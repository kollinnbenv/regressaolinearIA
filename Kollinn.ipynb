```python
import pandas as pd 
dataset = pd.read_csv("callcenter.csv")
```

```python
dataset.head()
#dataset.isnull().sum() ##so trata valor nulo em categorico 
```

```python
###separando valor numerico e categ√≥rico 
```

```python
dataset_categorico = dataset[['profissao', 'estado_civil', 'educacao', 'inadimplente', 'emprestimo_moradia', 'emprestimo_pessoal', 'meio_contato', 'mes', 'dia_da_semana', 'campanha_anterior']]
dataset_numerico = dataset[['idade', 'duracao', 'qtd_contatos_campanha', 'dias_ultimo_contato', 'qtd_contatos_total', 'indice_precos_consumidor', 'indice_confianca_consumidor', 'euribor3m']]
dataset_numerico.head()
```

```python
##tratando erros nulos 
```

```python
for nome_coluna in dataset_categorico.columns:
    dataset_categorico[nome_coluna].fillna('NULO', inplace=True)
```

```python
dataset_categorico.head()
```

```python
dataset_categorico_corrigido = pd.get_dummies(dataset_categorico)
dataset_categorico_corrigido.head()
```

```python
##juntando dados de treinamento
```

```python
x = pd.concat([dataset_categorico_corrigido, dataset_numerico], axis=1)
x.head()
```

```python
def categorize(valor):
    if valor == 'nao':
        return 0  #nao
    return 1  #sim
```

```python
y = dataset['resultado'].apply(categorize)
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
x_test.reset_index(drop=True, inplace=True)
y_test.reset_index(drop=True, inplace=True)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression() #Crio o Modelo
lr.fit(x_train,y_train) #Fa√ßo o Fit do Modelo (Reta)
previsto = lr.predict(x_test) #Predi√ß√£o nos Dados Teste
previsto
```
